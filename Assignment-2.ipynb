1. Create the pattern using nested for loop in Python. 
 

```python
for i in range(6):
    for j in range(i):
        print('*',end=" ") 
    if i ==5: 
        print("\r")
        for k in range(i-1,0,-1):
            for l in range(k):
                print("*", end=" ")
            print(" ")           
    print(" ")      
    
   
```

2. Write a Python program to reverse a word after accepting the input from the user. Sample Output: 
Input word: ineuron 


```python
input = 'ineuron'
```

```python
for i in range(len(input)):
    print(input[-i-1],end="") 
```
